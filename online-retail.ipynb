```python
import pandas as pd

df = pd.read_excel('Online Retail.xlsx')

display(df)
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
# Suppression des lignes sans CustomerID
df_retail = df.copy()
df_retail = df_retail.dropna(subset=['CustomerID'])
df_retail['CustomerID'] = df_retail['CustomerID'].astype('int64')
# Gestion des descriptions non renseignées
df_retail['Description'] = df['Description'].fillna('Unknown')
# Vérification
df_retail.isnull().sum()
```

```python
display(df_retail)
df_retail.describe()
```

```python
# Suppression des quantités négatives
df_retail = df_retail[df_retail['Quantity'] > 0]
df_retail[['Quantity']].describe()
```

```python
# les "vrais" produits ont un StockCode composé d'au moins 5 caractères
# donc nettoyage selon ce critère
df_retail = df_retail[df_retail['StockCode'].str.len() >= 5]
display(df_retail)
```

```python
# Suppression des lignes pour InvoiceNo commençant par 'C' (canceled invoices)
# Convert InvoiceNo column to string
df_retail['InvoiceNo'] = df_retail['InvoiceNo'].astype(str)
# Suppression des lignes pour InvoiceNo commençant par 'C' (canceled invoices)
df_retail = df_retail[~df_retail['InvoiceNo'].str.startswith('C')]
df_retail = df_retail[~df_retail['InvoiceNo'].str.startswith('c')]
display(df_retail)
```

```python
# Convert InvoiceNo column to int
df_retail['InvoiceNo'] = df_retail['InvoiceNo'].astype(int)
df_retail.info()
```

```python
# Création de la colonne TotalAmount = Quantity * UnitPrice
df_retail['TotalAmount'] = df_retail['Quantity'] * df_retail['UnitPrice']
display(df_retail)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grouper par date pour les ventes totales
df_daily = df_retail.groupby(df_retail['InvoiceDate'].dt.date)['TotalAmount'].sum()

# Série chronologique
df_daily.plot(kind='line', figsize=(10, 6))
plt.title('Évolution des ventes journalières')
plt.ylabel('Montant Total des Ventes')
plt.show()
```

```python
# Top 10 des pays avec le plus de ventes
country_sales = df_retail.groupby('Country')['TotalAmount'].sum().nlargest(10)

country_sales.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Top 10 des pays par ventes')
plt.ylabel('Montant Total des Ventes')
plt.show()
```

```python
# Filtrer les données pour ne conserver que le Royaume-Uni
df_uk = df_retail[df_retail['Country'] == 'United Kingdom']

display(df_uk)
```

```python
# Top 10 des clients
customer_sales = df_uk.groupby('CustomerID')['TotalAmount'].sum().nlargest(10)

customer_sales.plot(kind='bar', figsize=(10, 6), color='green')
plt.title('Top 10 des clients par ventes')
plt.ylabel('Montant Total des Ventes')
plt.show()
```

```python
# Définir une date de référence (par exemple, la dernière date dans les données)
reference_date = df_uk['InvoiceDate'].max()

# Calcul des métriques RFM
rfm = df_uk.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,  # Recency
    'InvoiceNo': 'count',  # Frequency
    'TotalAmount': 'sum'   # Monetary
}).rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalAmount': 'Monetary'
})

# Vérifier les résultats
display(rfm)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Recency vs Frequency
sns.scatterplot(data=rfm, x='Recency', y='Frequency', hue='Monetary', size='Monetary', sizes=(20, 200), palette='viridis')
plt.title('Recency vs Frequency (avec Monetary comme couleur)')
plt.xlabel('Recency (jours)')
plt.ylabel('Frequency')
plt.show()

# Frequency vs Monetary
sns.scatterplot(data=rfm, x='Frequency', y='Monetary', hue='Recency', size='Recency', sizes=(20, 200), palette='viridis')
plt.title('Frequency vs Monetary (avec Recency comme couleur)')
plt.xlabel('Frequency')
plt.ylabel('Monetary')
plt.show()

# Recency vs Monetary
sns.scatterplot(data=rfm, x='Monetary', y='Recency', hue='Frequency', size='Frequency', sizes=(20, 200), palette='viridis')
plt.title('Frequency vs Monetary (avec Frequency comme couleur)')
plt.xlabel('Frequency')
plt.ylabel('Monetary')
plt.show()
```

```python
# Catégorisation des valeurs de Recency, Frequency, Monetary
rfm['R_Score'] = pd.qcut(rfm['Recency'], q=4, labels=[4, 3, 2, 1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'], q=4, labels=[1, 2, 3, 4])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], q=4, labels=[1, 2, 3, 4])

# Créer une heatmap : Recency et Frequency
heatmap_data = rfm.groupby(['R_Score', 'F_Score'], observed=True)['Monetary'].mean().unstack()
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap='coolwarm')
plt.title('Heatmap - Recency vs Frequency (Monetary moyen)')
plt.xlabel('Frequency Score')
plt.ylabel('Recency Score')
plt.show()
```

```python
# Créer une heatmap : Recency et Monetary
heatmap_data = rfm.groupby(['R_Score', 'M_Score'], observed=True)['Frequency'].mean().unstack()
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap='coolwarm')
plt.title('Heatmap - Recency vs Monetary (Frequency moyen)')
plt.xlabel('Monetary Score')
plt.ylabel('Recency Score')
plt.show()
```

```python
# Créer une heatmap : Frequency et Monetary
heatmap_data = rfm.groupby(['M_Score', 'F_Score'], observed=True)['Recency'].mean().unstack()
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap='coolwarm')
plt.title('Heatmap - Frequency vs Monetary (Recency moyen)')
plt.xlabel('Monetary Score')
plt.ylabel('Frequency Score')
plt.show()
```

```python
import numpy as np

# Définir les seuils pour RFM
r_bins = [0, 30, 90, np.inf]
f_bins = [1, 2, 5, np.inf]
m_bins = [0, 500, 2000, np.inf]

# Assigner les scores RFM
rfm['R_Score'] = pd.cut(rfm['Recency'], bins=r_bins, labels=[3, 2, 1], include_lowest=True)
rfm['F_Score'] = pd.cut(rfm['Frequency'], bins=f_bins, labels=[1, 2, 3], include_lowest=True)
rfm['M_Score'] = pd.cut(rfm['Monetary'], bins=m_bins, labels=[1, 2, 3], include_lowest=True)

# Segmenter les clients
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

# Distribution des segments
rfm['RFM_Score'].value_counts().plot(kind='bar', figsize=(10, 6), color='orange')
plt.title('Distribution des segments RFM')
plt.xlabel('Segment RFM')
plt.ylabel('Nombre de clients')
plt.show()
```
